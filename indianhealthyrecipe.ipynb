```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv("IndianHealthyRecipe.csv")
dataset.head()
```

```python
dataset["Spice"].fillna(dataset["Spice"].mode()[0], inplace=True)
dataset["Prep Time"].fillna(dataset["Prep Time"].mode()[0], inplace=True)
dataset["Number of Votes"].fillna(dataset["Number of Votes"].median(), inplace=True)
dataset["Serves"].fillna(dataset["Serves"].mode(), inplace=True)
dataset["Cook Time"].fillna(dataset["Cook Time"].mode()[0], inplace=True)
```

```python
dataset.info()
```

After fillna method we fill null values with mod or median. For numbers(int , float types) we use mod , and for texts (string types) we use median.

```python
lst_str1 =[]
str1 = ' '
for time1 in dataset["Prep Time"]:
  str1= time1
  lst_str1.append(str1)
```

```python
lst1 = []
import re
for st1 in range(len(lst_str1)):
  prep_time = re.findall(r'\d+' , lst_str1[st1])
  lst1.append(prep_time)
```

```python
dataset["Prep Time"]=lst1
```

```python
for h in range(len(dataset["Prep Time"])):
  if len(dataset["Prep Time"][h])>1:
    hrs = int(dataset["Prep Time"][h][0])*60 + int(dataset["Prep Time"][h][1])
    dataset["Prep Time"][h] = hrs
```

```python
lst_str2 =[]
for time2 in dataset["Cook Time"]:
  str2= time2
  lst_str2.append(str2)
```

```python
lst2 = []
import re
for st2 in range(len(lst_str2)):
  cook_time = re.findall(r'\d+' , lst_str2[st2])
  lst2.append(cook_time)
```

```python
dataset["Cook Time"] = lst2
```

```python
for min in range(len(lst2)):
  if lst2[min][0]=='1':
    dataset["Cook Time"][min]= int(lst2[min][0])*60
    if len(lst2[min])>1:
      dataset["Cook Time"][min] = int(lst2[min][0])*60  + int(lst2[min][1])
```

```python
categories = ["Spice", "Prep Time", "Dietary Info", "Cook Time"]
fig1, axes1 = plt.subplots(nrows=1, ncols=4, figsize=(15, 7))
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 10))
for i,j in enumerate(categories):
    counts = df[j].value_counts()
    counts.plot(kind="bar", ax=axes1[i])
    for container in axes1[i].containers:
        axes1[i].bar_label(container)
    axes[i].pie(counts, autopct="%0.2f%%")
    axes[i].legend(counts.index)

plt.tight_layout()
plt.show()
```

```python
crosstabs = dict()

for i, j in enumerate(categories[:-1]):
    for k in categories[i+1:]:
        crosstab = pd.crosstab(df[j], df[k])
        crosstabs["{} vs {}".format(j, k)] = crosstab
```

```python
keys = [i for i in crosstabs.keys()]
index = 0
for i in range(2):
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))
    for j in range(3):
        sns.heatmap(crosstabs[keys[index]], ax=axes[j], annot=True, fmt="d")
        axes[j].set_title(keys[index])
        index += 1
    plt.tight_layout()
    plt.show()
```
